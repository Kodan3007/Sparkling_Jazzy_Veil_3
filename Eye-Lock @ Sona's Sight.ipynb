```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('eye_lock_mind_block.csv')
```

```python
plt.figure(figsize=(9,7))
plt.scatter(data['Longitude(Her Eye Depth)'], data['Latitude(My Look Intensity)'], 
                        c=range(len(data)),cmap='plasma', s=140, alpha=0.85,
            edgecolors='black')
for i, row in data.iterrows():
    plt.text(row['Longitude(Her Eye Depth)'] + 0.015, 
             row['Latitude(My Look Intensity)'] + 0.005,
             row['Emotion'], fontsize=8, color='black')
plt.xlabel("Her Eye Depth >", fontsize=12)
plt.ylabel("My Look Intensity >", fontsize=12)
plt.title("20 Moments of Eye Contact", fontsize=15, pad=20)
plt.grid(True, linestyle='--',
         alpha=0.3)
plt.show()
```
